```python
titles=artists.columns.tolist()
```

```python
artists=artists.T
artists.shape
```

```python
from sklearn.decomposition import NMF

model = NMF(n_components=3)

model.fit(artists)

nmf_features = model.transform(artists)

print(nmf_features.shape)

```

```python
df=pd.DataFrame(nmf_features,index=titles)
df
```
