# Recommendation System using above Sparse data

```python
# STEP 1 : We have to create nmf_features using NMF model(data should be sparse)
```

```python
from sklearn.decomposition import NMF

model=NMF(n_components=6)
model.fit(df)
nmf_features=model.transform(df)
nmf_features.shape
```

```python
artists= artists['Artist'] # convert artists to list
```

```python
# STEP 2: normalize the nmf_features and after normalizing create a DataFrame and give index = artists(or that thing on which you want recommendations)
from sklearn.preprocessing import normalize

norm_features = normalize(nmf_features)
df1 = pd.DataFrame(norm_features,index = artists)
artist = df1.loc['Sublime']
similarities = df1.dot(artist)

print(similarities.nlargest())

```
