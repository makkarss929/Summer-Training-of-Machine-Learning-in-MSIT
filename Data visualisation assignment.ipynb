```python
data['quality'].unique()
```

```python
new_quality=[]
for e in data['quality'].values:
    if e==3 or e==4:
        new_quality.append('poor')
    elif e==5 or e==6:
        new_quality.append('average')
    elif e==7 or e==8:
        new_quality.append('good')        
```

```python
data['new quality']=new_quality
```

```python
data['new quality'].unique()
```

```python
sns.lmplot(x='density',y='alcohol',hue='new quality',data=data,ci=68)
plt.show()
```

```python
new_residual_sugar=[]
for e in data['residual sugar'].values:
    if e<=4:
        new_residual_sugar.append('low')
    else:
        new_residual_sugar.append('high')
```

```python
data['new residual sugar']=new_residual_sugar
```
