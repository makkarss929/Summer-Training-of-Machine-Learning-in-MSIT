# PCA for titanic dataset

```python
data1=data.drop('Survived',axis=1)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

data1=StandardScaler().fit_transform(data1)
model=PCA()
transformed_data=model.fit_transform(data1)

variances = model.explained_variance_
print(variances)
features = range(len(variances))

plt.bar(features,variances)
plt.xticks(features)
plt.xlabel('features')
plt.ylabel('Variances')
plt.title("Variance of titanic features")
plt.show()
```

#### Now we have to remove low variance features 

```python
model = PCA(n_components=8)
pca_features = model.fit_transform(data1)
print(model.n_components_)
print(model.n_features_)
```

```python
model.components_
```

```python
pca_features
```

```python
new_features=np.matmul(data1,model.components_.T)
new_features
```

### Now new_features contain new data and now we are going to apply KMeans with new_features and predict the results by comparing with data['Survived']


```python
from sklearn.cluster import KMeans
```

```python
model= KMeans(n_clusters=2)
model.fit(new_features)
labels=model.predict(new_features)
labels

plt.scatter(new_features[:,0],new_features[:,1],c=labels)
```

```python
(labels==data['Survived']).sum()
```
