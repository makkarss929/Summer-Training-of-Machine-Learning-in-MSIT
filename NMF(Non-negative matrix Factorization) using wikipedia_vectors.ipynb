# Non-neagative Matrix Factorization(NMF)

```python
import pandas as pd
import numpy as np
```

```python
articles = pd.read_csv(r'C:\Users\AMIT\Desktop\MSIT_ML_CLASS-master-7\MSIT_ML_CLASS-master\datasets\wikipedia\wikipedia-vectors.csv')
```

```python
articles.head()
```

```python
articles.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
titles=list(articles.columns)   # or articles.columns.tolist()
```

```python
print(titles)
```

```python
articles = articles.T
articles.shape
```

```python
# Import NMF
from sklearn.decomposition import NMF

# Create an NMF instance or example : model
model=NMF(n_components=6)

# Fit the model to articles
model.fit(articles)

# Transform the articles :nmf_features
nmf_features = model.transform(articles)

# print the NMF features 
print(nmf_features)

print(nmf_features.shape)
```

# Recommendation Using NMF
### Which articles are similar to 'Cristiano Ronaldo' ?

Use NMF features and the cosine similarity to find similar articles .Apply this to your NMF model for popular wikipedia
articles,by finding the articles most similar to the article about the Cristiano Ronaldo.

```python
# perform the neccessary imports
import pandas as pd
from sklearn.preprocessing import normalize

# Normalize the NMF features :norm_features
norm_features = normalize(nmf_features)

# Create a DataFrame :df
df=pd.DataFrame(norm_features,index=titles)  # norm_features contain 60 rows and 6 features

# Select the row corresponding to Cristiano Ronaldo'  # article contain 1 row from norm_features and 6 features
article = df.loc['Cristiano Ronaldo']

# Compute the dot products:similarities
similarities=df.dot(article)          # element wise product and sum and output is 1 by 6

# Display those with the largest cosine similarity
print(similarities.nlargest())       # this will recommend top 5 similarities
```

```python
print(similarities.nsmallest()) # this will give top 5 dissimalrities
```
