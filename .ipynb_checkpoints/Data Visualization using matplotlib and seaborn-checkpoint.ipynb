```python
plt.subplot(1,2,1)
# subplot three arguments 
#1st --> number of rows
#2nd --> number of columns
#3rd --> plot number

plt.plot(data['fixed acidity'])
plt.title('Fixed Acidity')
plt.xlabel('number of wines')


plt.subplot(1,2,2)
plt.plot(data['pH'])
plt.title('pH')

# for common title
plt.suptitle('Fixed Acidity and pH')
plt.show()
```

```python
# how to give different colours
plt.subplot(1,2,1)
plt.plot(data['fixed acidity'],color='red')
plt.title('Fixed Acidity')
plt.xlabel('number of wines')


plt.subplot(1,2,2)
plt.plot(data['pH'],color='blue')
plt.title('pH')

# for common title
plt.suptitle('Fixed Acidity and pH')
plt.show()
```

```python
plt.figure(figsize=(10,5)) # units in inches

plt.subplot(1,2,1)
plt.plot(data['fixed acidity'],color='red')
plt.title('Fixed Acidity')
plt.xlabel('number of wines')


plt.subplot(1,2,2)
plt.plot(data['pH'],color='blue')
plt.title('pH')

# for common title
plt.suptitle('Fixed Acidity and pH')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.jointplot(x='fixed acidity',y='pH',data=data)
plt.show()
```

```python
# linear model plot
sns.lmplot(x='fixed acidity',y='pH',data=data)
plt.show()
```

```python
# regression plot with order 1
sns.regplot(x='fixed acidity',y='pH',data=data,order=1)
plt.show()
```

```python
# regression plot with order 2
sns.regplot(x='fixed acidity',y='pH',data=data,order=2)
plt.show()
```

```python
# regression plot with order 3
sns.regplot(x='fixed acidity',y='pH',data=data,order=3)
plt.show()
```
